```python
amount = float(input("Enter the transaction amount: "))
if amount > 0:
    print("The transaction amount is Positive.")
elif amount < 0:
    print("The transaction amount is Negative.")
else:
    print("The transaction amount is Zero.")
```

```python
number = int(input("Enter your passcode (numeric): "))
digit_sum = 0
while number > 0:
    digit = number % 10       # get the last digit
    digit_sum += digit        # add it to the sum
    number //= 10             # remove the last digit
print("Sum of digits:", digit_sum)
```

```python
transaction_id = int(input("Enter the transaction ID: "))
reversed_id = 0
while transaction_id > 0:
    digit = transaction_id % 10          # extract last digit
    reversed_id = reversed_id * 10 + digit  # build reversed number
    transaction_id //= 10                # remove last digit
print("Reversed transaction ID:", reversed_id)
```

```python
num = int(input("Enter the user ID: "))
if num <= 1:
    print("Not a prime number.")
else:
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print("Prime number.")
    else:
        print("Not a prime number.")
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
num = int(input("Enter a number: "))
result = factorial(num)
print("Factorial of", num, "is:", result)

```

```python
num = int(input("Enter a number: "))
num_str = str(num)
num_digits = len(num_str)
armstrong_sum = sum(int(digit) ** num_digits for digit in num_str)
if armstrong_sum == num:
    print("Armstrong number.")
else:
    print("Not an Armstrong number.")
```

```python

password = input("Enter your password: ")
if len(password) > 1:
    swapped_password = password[-1] + password[1:-1] + password[0]
else:
    swapped_password = password
print("Strengthened password:", swapped_password)
```

```python
decimal_num = int(input("Enter a decimal number: "))
binary_num = bin(decimal_num)
print("Binary equivalent:", binary_num[2:])
```

```python

sentence = input("Enter a sentence: ")
words = sentence.split()
longest_word = max(words, key=len)
print("The longest word is:", longest_word)
```

```python

str1 = input("Enter first string: ").lower()
str2 = input("Enter second string: ").lower()

str1 = str1.replace(" ", "")
str2 = str2.replace(" ", "")

if sorted(str1) == sorted(str2):
    print("The strings are anagrams.")
else:
    print("The strings are not anagrams.")
```
